```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt='.2f', cmap='crest', linewidths=0.25, linecolor='white', robust=True)
plt.title('Correlación entre variables', fontsize=13, fontweight='bold')
plt.show()
```

```python
subset = df[(df['Pais'].isin(['JAPÓN', 'CHINA', 'COREA DEL SUR', 'ESTADOS UNIDOS'])) & (df['price'] < 100_000_000)]

# get average price per year
subset = subset.groupby(['year', 'Pais'])['price'].mean().reset_index()

plt.figure(figsize=(10, 8))
sns.lineplot(x='year', y='price', hue='Pais', data=subset, alpha=0.7, palette=['red', 'grey', 'grey', 'grey'], style='Pais', markers=True, dashes=False, linewidth=1.5, markersize=4)
plt.title('Comparación de depreciación por país de origen', fontsize=13, fontweight='bold')
plt.xlabel('Año', fontsize=12)
plt.ylabel('Precio promedio', fontsize=12)

# add country labels to the end of each line
for pais in subset['Pais'].unique():
	x = subset[subset['Pais'] == pais]['year'].iloc[-1]
	y = subset[subset['Pais'] == pais]['price'].iloc[-1]
	plt.text(x, y, pais, fontsize=12, color='black' if pais == 'CHINA' else 'grey', ha='left', va='center')

# hide the legend
plt.legend().remove()

plt.show()
```

```python
subset = df[(df['Pais'].isin(['JAPÓN', 'CHINA', 'COREA DEL SUR', 'ESTADOS UNIDOS'])) & (df['price'] < 100_000_000)]

# get average price per year
km_intervals = [
	0, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000
]

subset['km'] = pd.cut(subset['km'], bins=km_intervals, labels=km_intervals[:-1])
subset = subset.groupby(['km', 'Pais'])['price'].mean().reset_index()

plt.figure(figsize=(10, 8))
sns.lineplot(x='km', y='price', hue='Pais', data=subset, alpha=0.7, palette=['red', 'grey', 'grey', 'grey'], style='Pais', markers=True, dashes=False, linewidth=1.5, markersize=4)
plt.title('Comparación de depreciación por país de origen', fontsize=13, fontweight='bold')
plt.xlabel('KM', fontsize=12)
plt.ylabel('Precio promedio', fontsize=12)

# add country labels to the end of each line
for pais in subset['Pais'].unique():
	x = subset[subset['Pais'] == pais]['km'].iloc[-1]
	y = subset[subset['Pais'] == pais]['price'].iloc[-1]
	plt.text(x, y, pais, fontsize=12, color='black' if pais == 'CHINA' else 'grey', ha='left', va='center')

# hide the legend
plt.legend().remove()

plt.show()
```

```python
# identificar modelos con mayor publicaciones por cada marca
marcas = df['Marca'].unique()

for marca in marcas:
	subset = df[df['Marca'] == marca]
	subset = subset.groupby(['Modelo'])['price'].count().reset_index()
	subset = subset.sort_values(by='price', ascending=False).head(5)

	print(f'Modelos más publicados de {marca}')
	print(subset)
```

```python
tipos = df['Tipo'].unique()

for tipo in tipos:
	modelo_top = df[df['Tipo'] == tipo].groupby(['Modelo']).count().reset_index().sort_values(by='price', ascending=False).head(5)
	subset = df[df['Modelo'].isin(modelo_top['Modelo'])]

	# comparación entre tipos de autos y precios en scatterplot
	plt.figure(figsize=(10, 8))
	sns.lmplot(x='km', y='price', hue='Modelo', palette='muted', data=subset, height=8, aspect=1, scatter_kws={'alpha': 0.6, 's': 15}, line_kws={'linewidth': 1.5}, ci=None)
	plt.title(f'Comparación de precios por tipo de vehículo - {tipo}', fontsize=13, fontweight='bold')
	plt.xlabel('KM', fontsize=12)
	plt.ylabel('Precio', fontsize=12)
	# adjust x axis limits
	plt.xlim(0, 200_000)
	plt.ylim(0, 30_000_000)

	plt.savefig(f'plots/TOP_VENTAS_TIPO_{tipo}.png', dpi=300, bbox_inches='tight')
	plt.show()
```
